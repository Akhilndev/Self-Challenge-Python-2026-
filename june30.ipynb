Below using some string methods to see how it works !

```python
Sidt = "akhl nemala"
myname  = Sidt.upper() # stringname ( example "S" here) ---- Stringname.upper()
print(myname)

myname = Sidt.title()
print(myname) # Title, Capitalize --- gives first letter capital for every word.

""" Complex methods to understand """

text = "Python"
print(text.startswith("Py")) #Counts how many times something appears.

text = "photo.jpg"
print(text.endswith(".jpg")) # Endswith -- checks- true or not
text = "Python"
print(text.startswith("Py")) # Starswith -- checks -  false or not
    
text = "Python"
print(text.index("th")) # Works exactly like find() except:

text = "Python Programming"
print(text.find("Pro")) #Returns the position (index) where the substring first appears.

""" 
Method	Not Found

find()	Returns -1
index()	Raises ValueErr
"""

#Join string ---> is like concentation (+) but it helps to join multiple strings in a list  to one single string
# placing the specified separator between them. !

fruits = ["Apple", "Banana", "Mango"]
result = ", ".join(fruits)
print(result)
```
